```python
if 25 < 10:
  print('It Worked!')
else:
  print('It did not work...')
```

```python
if 25 < 10:
  print('It Worked!')
elif 25 < 30:
  print('elift worked!')
elif 25 < 20:
  print('elift 2 worked!')
elif 25 < 40:
  print('elift 3 worked!')
elif 25 < 50:
  print('elift 4 worked!')
else:
  print('It did not work...')
```
