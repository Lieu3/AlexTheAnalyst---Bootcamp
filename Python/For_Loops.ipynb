```python
integers = [1,2,3,4,5]
```

```python
for number in integers:
  print(number)
```

```python
integers = [1,2,3,4,5]
```

```python
for Jelly in integers:
  print(Jelly + Jelly)
```

```python
ice_cream_dict = {'name': 'Blake L', 'weekly intake':5,'favorite ice cream':['Mint','Chocolate']}
```

```python
for cream in ice_cream_dict.values():
  print(cream)
```

```python
for key, value in ice_cream_dict.items():
  print (key, '->',value)
```

```python
#Nested For Loops
```

```python
flavors = ['Vanilla','Chocolate','Cookie Dough']
toppings = ['Hot Fudge','Oreos','Marshmallows']
```

```python
for one in flavors:
  for two in toppings:
    print(one, 'topped with', two)
```
