```python
number = 0
while number < 5:
  print(number)
  number = number + 1
```

```python
#Break statment

number = 0
while number < 5:
  print(number)
  if number == 3:
    break
  number = number + 1
```

```python
#Else statment

number = 0
while number < 5:
  print(number)
  if number == 6:
    break
  number = number + 1
else:
  print('No Longer < 5')
```

```python
# Continue Statement

number = 0
while number < 5:
  print(number)
  if number == 3:
    continue
  number = number + 1
else:
  print('No Longer < 5')

#this will repeat the output 3
```

```python
number = 0

while number < 5:
  number = number + 1
  if number == 3:
    continue
  print(number)
else:
  print('No Longer < 5')

#this skips 3 but continues
```
